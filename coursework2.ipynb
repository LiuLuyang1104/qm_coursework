```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy as sp

import scipy.stats as sps
```

```python
# Load the data into data frames

d1 = pd.read_csv('gcse_results_clear.csv')
```

```python
plt.style.use('seaborn-white')
x2 = d1["2019_a8_b"]
x3 = d1["2019_a8_g"]
kwargs = dict(histtype='stepfilled', alpha=0.3, bins=10)
plt.hist(x2, **kwargs,label="Boys")
plt.hist(x3, **kwargs,label="Girls",color='r')
plt.xlabel("Average Progress 8 score per pupil",fontsize="large")
plt.ylabel("Number of areas",fontsize="large")
plt.legend()
plt.title('Average scores of different genders in 2019',fontsize='x-large',fontweight='bold') 
```
